####**IDENTIFY SALES TRENDS, PATTERNS, AND SEASONALITY USING GRAPHS**.

```python
# Convert order date to date time
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = True)
```

```python
# Create a time series by month
df['Order Month'] = df['Order Date'].dt.to_period('M').astype(str)
```

```python
monthly_sales = df.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
# Sales trends over time
plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_sales, x='Order Month', y='Sales', marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#####**Key insights:**


The chart displays clear monthly fluctuations in total sales across 2019 and 2020.

Sales show periodic peaks, especially during:

November and December 2020 – the highest sales, likely due to year-end holiday promotions.

Smaller spikes in March, September, and November 2019 suggest seasonal or campaign-based boosts.

The overall trend is upward, indicating improved performance over time despite fluctuations.

```python
# additional features for analysis
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()

# Aggregate sales by month across all years
monthly_avg_sales = df.groupby('Month')['Sales'].mean().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]).reset_index()

# Aggregate yearly sales
yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()

plt.figure(figsize=(14, 6))
sns.lineplot(data=yearly_sales, x='Year', y='Sales', marker='o')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Key Insights:**

The Yearly Sales Trend chart shows how total sales changed each year.

The chart displays a steady and significant increase in yearly sales from the previous year to the next.

```python
# Monthly average sales (seasonality)
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_avg_sales, x='Month', y='Sales', palette='coolwarm')
plt.title('Average Sales by Month (Seasonality)')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

**Key Insights:**

February has the highest average sales, which may result from fiscal year-end purchases, corporate budgeting cycles, or seasonal promotions.

December and January also show strong performance, aligning with holiday season spending and year-start planning.

April to September generally see lower average sales, suggesting a mid-year dip in customer spending or reduced promotional activity.


```python
# 3. Sales by Product Category
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis')
plt.title('Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

**Key Insight from "Sales by Product Category"**

The bar chart highlights how sales are distributed across different product categories. From this chart, the key insights are:

**Office Supplies** generate the highest total sales, indicating strong and frequent demand—possibly due to low-cost, high-volume purchases.

**Technology** ranks second, showing strong revenue—likely from higher-priced items like electronics.

**Furniture** contributes the least, suggesting either lower sales volume or less frequent purchases.

```python
# Sales by Region
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(6, 4))
sns.barplot(x=region_sales.index, y=region_sales.values, palette='viridis')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
```

**Key Insight from "Sales by Region"**

The West region leads with the highest total sales, suggesting strong customer demand or a well-established market presence.

East and Central regions follow, contributing moderately to overall revenue.

The South region shows the lowest sales, indicating potential underperformance, limited market reach, or lower demand.

#### **PRESENT KEY KPIS SUCH AS TOTAL SALES, REVENUE PER CATEGORY, AND AVERAGE ORDER VALUE**.

**Total Sales:**

Indicates the overall revenue generated. It reflects the scale and financial performance of the business during the recorded period.

**Revenue by Category:**

Office Supplies is the top-selling category, followed by Technology and Furniture.

**Average Order Value(AOV):**

A key metric for understanding customer spending behavior. A rising AOV often indicates success in upselling or effective bundling strategies

These metrics provide a quick, high-level overview of business performance and are essential for tracking financial health, planning marketing strategies, and setting sales goals.
